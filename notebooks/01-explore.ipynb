```python
import pandas as pd
import numpy as np
import os
os.listdir('.')
```

```python
df_test = pd.read_csv("../data/raw/test.csv")
df_test.head()
```

```python
import json

def convert_to_dict(s):
    if isinstance(s, str):
        return json.loads(s)
    elif isinstance(s, dict):
        return s
    else:
        raise
        return

def read_json_col(df, col, key):
    return df[col].apply(lambda x: convert_to_dict(x)[key])
```

```python
JSON_COLS = 'device geoNetwork totals trafficSource'.split()

for col in JSON_COLS:
    tmp = json.loads(df_test[col][0])
    for key in tmp.keys():
        try:
            df_test[key] = read_json_col(df_test, col, key)
        except Exception as ex:
            print('Error: col={}, key={}'.format(col, key))
            print(ex)
            pass
    df_test.drop(labels=col, axis=1, inplace=True)
```

```python
for key in df_test['adwordsClickInfo'][0].keys():
    df_test[key] = read_json_col(df_test, 'adwordsClickInfo', key)
    df_test.drop(labels='adwordsClickInfo', axis=1, inplace=True)
```
